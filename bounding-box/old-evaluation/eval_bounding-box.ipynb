### Data Retrieval

```
vid_id = 'VIRAT_S_000201_08_001652_001838'
```

```
# unpickle files
yolo_data = open('detect/' + vid_id + '-YOLO_results.pkl', 'rb')
classID_list, scores_list, boundingbox_list = pickle.load(yolo_data)
detr_data = open('detect/' + vid_id + '-DETR_results.pkl', 'rb')
probas_list, bbox_list = pickle.load(detr_data)
```

### YOLO Results/Evaluation


```
# count violations of social distancing given list of bounding boxes
count_violations = []
for idx, boxes in enumerate(boundingbox_list):
  # print every 100 entries
  if idx % 100 == 0:
    print("frame #:", idx)
  # run count_undistanced
  num_pairs, pair_ids, pairwise_distance_array = box.count_undistanced(boxes.asnumpy())
  count_violations.append(num_pairs)
```

```
print("Total # frames:", len(boundingbox_list))
print("# of occurrences of no social distancing:", sum(filter(None, count_violations)))
```

```
# replace None values with numpy nan
for idx, entry in enumerate(count_violations):
  if entry == None:
    count_violations[idx] = np.nan
count_violations = np.array(count_violations)
```

```
np.unique(count_violations[~np.isnan(count_violations)])
```

```
index_dict = {}
for val in np.unique(count_violations[~np.isnan(count_violations)]):
  index_dict[val] = np.where(count_violations==val)[0]
index_dict[np.nan] = np.where(np.isnan(count_violations))[0]
```

```
index_dict
```

### DETR Results/Evaluation

```
# count violations of social distancing given list of bounding boxes
count_violations = []
for idx, boxes in enumerate(bbox_list):
  # print every 100 entries
  if idx % 100 == 0:
    print("frame #:", idx)
  # only get those entries that are people
  person_boxes = []
  for p, box_entry in zip(probas_list[idx], boxes):
    cl = p.argmax()
    if cl == 1:
      person_boxes.append(box_entry)
  # run count_undistanced
  num_pairs, pair_ids, pairwise_distance_array = box.count_undistanced(person_boxes)
  count_violations.append(num_pairs)
```

```
print("Total # frames:", len(bbox_list))
print("# of occurrences of no social distancing:", sum(filter(None, count_violations)))
```

```
# replace None values with numpy nan
for idx, entry in enumerate(count_violations):
  if entry == None:
    count_violations[idx] = np.nan
count_violations = np.array(count_violations)
```

```
np.unique(count_violations[~np.isnan(count_violations)])
```

```
index_dict = {}
for val in np.unique(count_violations[~np.isnan(count_violations)]):
  index_dict[val] = np.where(count_violations==val)[0]
index_dict[np.nan] = np.where(np.isnan(count_violations))[0]
```
