### Create Made-Up Intrinsic Matrix, K

```python
def get_intrinsic(image, f=10):
    """ Get intrinsic matrix given image and arbitrary focal length parameter f."""
    # get f in terms of fx and fy
    if isinstance(f, int):
        fx = f
        fy = f
    elif len(f)==1:
        fx = f[0]
        fy = f[0]
    elif len(f)!=2:
        print("Error in length of f: len =", len(f))
        return
    else:
        fx = f[0] 
        fy = f[1] 
    # center of image is principal point (ox, oy)
    ox = (image.shape[0]+1)/2
    oy = (image.shape[1]+1)/2
    # get intrinsic matrix K
    K = np.array([[fx,0,ox],[0,fy,oy],[0,0,1]])
    return K
```

```python
K = get_intrinsic(frame0)
print(K)
```

### Decompose Rotation Matrix

```python
def decompose_rotation(rot_matrix):
    """ Decompose rotation matrix into 3 Euler angles. """
    theta_x = np.arctan2(rot_matrix[2,1], rot_matrix[2,2])
    theta_y = np.arctan2(-rot_matrix[2,0], np.sqrt(np.square(rot_matrix[2,1])+np.square(rot_matrix[2,2])))
    theta_z = np.arctan2(rot_matrix[1,0], rot_matrix[0,0])
    return theta_x, theta_y, theta_z
```

```python
thetas = []
for idx, R in enumerate(np.unique(Rs, axis=0)):
    theta_x, theta_y, theta_z = decompose_rotation(R)
    print("unique rotation matrix #%d" % (idx+1))
    print(R)
    print("theta x y z:", theta_x, theta_y, theta_z, "\n")
    thetas.append([theta_x, theta_y, theta_z])
```
